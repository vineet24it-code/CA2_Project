```python
#Data Collection and Understanding
#Loading CSV
import pandas as pd
df = pd.read_csv("/content/gold_price_forecasting_dataset.csv")
print(df)
```

```python
df.head()
df.info()
df.describe()
```

```python
#Data Cleaning Techniques
#Handling Missing Values
#Mean
df.fillna(df.mean(numeric_only=True), inplace=True)
print(df.mean(numeric_only=True))
```

```python
#median
df.fillna(df.median(numeric_only=True), inplace=True)
print(df.median(numeric_only=True))
```

```python
#drop
df.dropna(inplace=True)
print(df)
```

```python
#Handling Duplicate Records
df.drop_duplicates(inplace=True)
print(df)
```

```python
#Detecting Outliers
import matplotlib.pyplot as plt
plt.boxplot(df['close'])
plt.title("Outlier Detection using Box Plot")
plt.show()
```

```python
#Renaming Columns
df.rename(columns={'close':'Gold_Close_Price'}, inplace=True)
print(df.Gold_Close_Price)
```

```python
#Data Transformation Techniques
#Datatype Conversion
df['date'] = pd.to_datetime(df['date'])
print(df.dtypes)
```

```python
#Scaling Data
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[['Gold_Close_Price']] = scaler.fit_transform(df[['Gold_Close_Price']])
print(df[['Gold_Close_Price']])
```

```python
#Normaliztion
df['Gold_Close_Price'] = (df['Gold_Close_Price'] - df['Gold_Close_Price'].mean()) / df['Gold_Close_Price'].std()
print(df['Gold_Close_Price'])
```

```python
#Binning
df['Price_Category'] = pd.cut(df['Gold_Close_Price'], bins=3, labels=['Low','Medium','High'])
print(df['Price_Category'])
```

```python
#Aggregation Functions
df.groupby(df['date'].dt.year)['Gold_Close_Price'].mean()
```

```python
#Exploratory Data Analysis (EDA)
#Mean, Median, Mode
df['Gold_Close_Price'].mean()
df['Gold_Close_Price'].median()
df['Gold_Close_Price'].mode()
```

```python
#Visualizations
#Histogram
import matplotlib.pyplot as plt

plt.hist(df['Gold_Close_Price'])
plt.title("Gold Price Distribution")
plt.show()
```

```python
#Box Plot
plt.boxplot(df['Gold_Close_Price'])
plt.show()
```

```python
#Line Chart
plt.plot(df['date'], df['Gold_Close_Price'])
plt.show()
```

```python
#Scatter Plot
plt.scatter(df.index, df['Gold_Close_Price'])
plt.show()
```

```python
#One-Hot Encoding
pd.get_dummies(df['Price_Category'])
```

```python
#Dummy Variable Creation
df_dummies = pd.get_dummies(df['Price_Category'], drop_first=True)
print(df_dummies.head())

```

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Price_Category'] = le.fit_transform(df['Price_Category'])
print(df['Price_Category'])
```

```python
#Feature Extraction
df['Year'] = df['date'].dt.year
df['Month'] = df['date'].dt.month
df['Day'] = df['date'].dt.day
print(df[['Year','Month','Day']])
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[['Gold_Close_Price']] = scaler.fit_transform(df[['Gold_Close_Price']])
print(df[['Gold_Close_Price']])
```

```python
#Dimensionality Reduction
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
reduced_data = pca.fit_transform(df.select_dtypes(include='number'))
print(reduced_data)
```

```python
#Text Data Processing Techniques
#Lowercase Conversion
df['Text'] = 'Gold_Close_Price daily_return'
df['Text'] = df['Text'].str.lower()
print(df['Text'].head())
```

```python
#Removing Punctuation
import string

df['Text'] = df['Text'].str.replace(f"[{string.punctuation}]", "", regex=True)
print(df['Text'].head())

```

```python
#Tokenization
df['Tokens'] = df['Text'].str.split()
print(df['Tokens'].head())

```
